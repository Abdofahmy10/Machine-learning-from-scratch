```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from Linear_regression_from_scatch import LinearRegression
```

```python
X,y = datasets.make_regression(n_samples=100,n_features=1,noise=20 , random_state=42)
```

```python
X_train ,X_test,y_train ,y_test = train_test_split(X,y ,test_size= 0.2 ,random_state= 142)
```

```python
y_train.shape
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0],y ,color = 'b',s = 30)
```

```python
LR = LinearRegression(lr=0.01)
```

```python
y_train.shape
```

```python
LR.fit(X_train , y_train) 
```

```python
y_hat = LR.predict(X_test)
```

```python
def mse (y_test, y_hat):
    return np.mean((y_test - y_hat)**2)
```

```python
mse = mse (y_test, y_hat)
```

```python
y_pred_line = LR.predict(X)
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
m1 = plt.scatter(X_train, y_train, color=cmap(0.9), s=10)
m2 = plt.scatter(X_test, y_test, color=cmap(0.1), s=10)
plt.plot(X, y_pred_line, color='black', linewidth=2, label='Prediction')
plt.show()
```
